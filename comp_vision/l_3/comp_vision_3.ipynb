```python
# Імпорт і завантаження зображення
import cv2
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = [15, 10]

img = cv2.imread("C:/Users/User/Downloads/11.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Original")
plt.axis(False)
```

```python
# Створення розмитого зображення (Gaussian blur)

unsharp = cv2.GaussianBlur(img, (9,9), 0)

plt.imshow(unsharp)
plt.title("Blurred")
plt.axis(False)
```

```python
# Переведення у float32
original_f = img.astype(np.float32)
unsharp_f = unsharp.astype(np.float32)
```

```python
# Різниця зображень
# difference = original − unsharp
diff = original_f - unsharp_f
```

```python
# Unsharp Masking
# amount — коефіцієнт різкості
amount = 1.5

sharpened = original_f + diff * amount
```

```python
# Clipping + повернення в uint8
sharpened[sharpened < 0] = 0
sharpened[sharpened > 255] = 255

sharpened = sharpened.astype(np.uint8)
```

```python
# Вивід результатів
plt.figure()

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Original")
plt.axis(False)

plt.subplot(1,3,2)
plt.imshow(unsharp)
plt.title("Blurred")
plt.axis(False)

plt.subplot(1,3,3)
plt.imshow(sharpened)
plt.title("USM Result")
plt.axis(False)

plt.show()
```

**Висновки**

У роботі було реалізовано фільтр нерізкого маскування (Unsharp Masking), який підвищує різкість зображення шляхом додавання різниці між оригінальним та розмитим зображенням.
Оптимальне значення параметра amount знаходиться приблизно в межах 1–2, оскільки при таких значеннях зображення стає більш чітким без появи сильних спотворень.

На основі отриманих результатів можна зробити висновок, що фільтр Unsharp Masking ефективно покращує чіткість зображення, але потребує правильного підбору параметрів, інакше може призвести до погіршення якості.
