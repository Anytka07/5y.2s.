```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Завантаження зображення
img = cv2.imread(r"C:\Users\User\Downloads\11.jpg")   # вкажіть назву свого файлу

# Ковертувати в RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Вивести зображення
plt.imshow(img)
plt.axis(False)
```

```python
# Розділення на три канали (червоний,зелений та синій)
red, green, blue = cv2.split(img)
```

```python
# Злиття в RGB
img1 = cv2.merge([red, green, blue])
```

```python
# Злиття в RBG
img2 = cv2.merge([red, blue, green])
```

```python
# Злиття в GRB
img3 = cv2.merge([green, red, blue])
```

```python
# Злиття в BGR
img4 = cv2.merge([blue, green, red])
```

```python
# Створення колажу
out1 = np.hstack([img1, img2])
out2 = np.hstack([img3, img4])
out = np.vstack([out1, out2])
```

```python
# Вивід колажу
plt.imshow(out)
plt.axis(False)
plt.show()
```

```python
## Висновки

1. При різних порядках каналів кольори на зображенні змінюються. Наприклад, деякі ділянки можуть ставати синіми, зеленими або червоними замість їх реального кольору.

2. Це відбувається тому, що кожен канал відповідає за інтенсивність певного кольору (Red, Green, Blue). Коли змінюється порядок каналів під час злиття, інформація одного кольору починає відображатися як інший.

3. У додатках комп'ютерного зору це може викликати проблеми, оскільки різні бібліотеки використовують різні порядки каналів. Наприклад, OpenCV використовує BGR, а matplotlib і більшість моделей машинного навчання використовують RGB. Якщо не виконати правильне перетворення, модель може отримати неправильні дані, що призведе до зниження точності розпізнавання або класифікації.
```
